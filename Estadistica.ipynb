```python
def factorial(numero):
    total = 1
    for i in range(1,numero+1):
        total = total * i
    return total
    
factorial(20)
```

```python
def binario(x):
    binario = ''
    while x > 0: 
        residuo = int(x % 2)
        x = int(x/2)
        binario = str(residuo) + binario
    return binario

num = int(input('Ingresa un número: '))
print(binario(num))
```
